### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv')
tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
tips.drop('Unnamed: 0', axis=1, inplace=True)
tips.head()
```

### Step 5. Plot the total_bill column histogram

```python
tips.total_bill.plot(kind='hist')
plt.xlabel('Total Bill Amount')
plt.ylabel('Frequency')
plt.title('Histogram of Total Bill Amount')
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.scatter(tips.total_bill, tips.tip)
plt.xlabel('Total Bill Amount')
plt.ylabel('Tip')
plt.title('Relationship Between Total Bill and Tip')
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
plt.scatter(tips.total_bill, tips.tip, c=tips['size'])
plt.xlabel('Total Bill Amount')
plt.ylabel('Tip Amount')
plt.title('Scatter Plot of Total Bill, Tip, and Size')
plt.colorbar(label='Size')
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.xlabel('Day')
plt.ylabel('Total Bill Amount')
plt.title('Box Plot of Total Bill Amount by Day')
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.scatterplot(data=tips,x='tip',y='day', hue='sex')
plt.xlabel('Tip Amount')
plt.ylabel('Day')
plt.title('Scatter Plot of Tip Amount by Day with Differentiated Sex')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(data=tips,x='day',y='total_bill', hue='time')
plt.xlabel('Day')
plt.ylabel('Total Bill Amount')
plt.title('Box Plot of Total Bill Amount by Day and Time')
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.histplot(tips[tips['time'] == 'Dinner']['tip'], color='blue')
plt.title('Dinner Tip Distribution')
plt.xlabel('Tip Amount')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(tips[tips['time'] == 'Lunch']['tip'], color='green')
plt.title('Lunch Tip Distribution')
plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(tips[tips['sex'] == 'Male']['total_bill'], color='blue')
plt.title('Male and Total Bill Distribution')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.scatterplot(tips[tips['sex'] == 'Female']['total_bill'], color='green')
plt.title('Female and Total Bill Distribution')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
```
